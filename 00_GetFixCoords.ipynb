```python
from __future__ import division
import csv
import urllib2
from bs4 import BeautifulSoup
import LatLon
import re
import time
```

```python
def FixCoords(Fixname,HEAD = 'Mozilla/5.0 (X11; U; Linux i686) Gecko/20071127 Firefox/2.0.0.11'):
    
    urllink = 'https://nfdc.faa.gov/nfdcApps/services/airspaceFixes/fix_search.jsp?selectType=state&selectName=AK&keyword='+Fixname
    req = urllib2.Request(urllink, headers={'User-Agent' : HEAD}) 
    response = urllib2.urlopen(req)
    html = response.read()
    soup = BeautifulSoup(html)
    table = soup.findAll('td')
    try:
        palmyra = LatLon.string2latlon(table[1].string.split('\n')[0], table[3].string.split('\n')[0], 'd%-%m%-%S% %H')
        Coords = palmyra.to_string()
        return [float(Coords[0]), float(Coords[1])]
    except:
        print Fixname
        return []
def NavaidCoords(Navaid, HEAD = 'Mozilla/5.0'):
    urllink = 'https://www.airnav.com/cgi-bin/navaid-info?' + Navaid
    req = urllib2.Request(urllink, headers={'User-Agent' : HEAD}) 
    response = urllib2.urlopen(req,timeout=10)
    html = response.read()
    soup = BeautifulSoup(html)
    table = soup.findAll('pre')
    try:
        DATA = table[0].string
        p = re.compile(r'[-+]?[0-9]*\.?[0-9]+/[-+]?[0-9]*\.?[0-9]+')
        Coords = p.findall(DATA)[0].split('/')
        return [float(Coords[0]), float(Coords[1])]
    except:
        print Navaid
        return []
```

```python
def NavaidCoords2(Navaid,HEAD = 'Mozilla/5.0'):
    urllink = 'http://worldaerodata.com/wad.cgi?search=%s&apt_nv='%Navaid
    req = urllib2.Request(urllink, headers={'User-Agent' : HEAD}) 
    response = urllib2.urlopen(req,timeout = 5)
    html = response.read()
    soup = BeautifulSoup(html)
    table = soup.findAll('table')
    try:
        Lat = float(table[4].findAll('td')[3].contents[0])
        Lon = float(table[4].findAll('td')[4].contents[0])
        return [Lat,Lon]
    except:
        print Navaid
        return []
```
