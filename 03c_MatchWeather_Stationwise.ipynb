## Location of valid stations

```python
Wx_stationwise_dir = os.getcwd() + '/NOAA/processed_stationwise/valid_station_wx_201301010000.csv'
Wx_station_dir = os.getcwd() + '/NOAA/WeatherStation.csv'
weather_station = pd.read_csv(Wx_station_dir)
weather_data_valid = pd.read_csv(Wx_stationwise_dir)
valid_station = pd.DataFrame(data = {'WBAN': weather_data_valid.WBAN.unique()}).merge(weather_station[['WBAN','LON','LAT']], how = 'inner')
valid_station = valid_station.sort_values(by = 'WBAN').reset_index(drop = 1)
```

```python
fig = plt.figure(figsize=(16,12))
m = Basemap(llcrnrlon = -128,llcrnrlat = 23, urcrnrlon = -63, urcrnrlat = 50,projection='merc')
m.bluemarble()
m.drawcoastlines(linewidth=0.5)
m.drawcountries(linewidth=0.5)
m.drawstates(linewidth=0.5)
m.drawparallels(np.arange(10.,35.,5.))
m.drawmeridians(np.arange(-120.,-80.,10.))

x, y = m(valid_station['LON'].values, valid_station['LAT'].values)
plt.scatter(x, y, color = 'r')
```
